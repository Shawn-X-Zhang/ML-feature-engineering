# First try with dataset with simple features (no if-then relationship between features)

```python
# Create dataset
import numpy as np
import pandas as pd
np.random.seed(1)
size = 1000
x1 = np.random.random(size)
x2 = np.random.random(size)
y = x1+x2+np.random.random(size)/100

dataset = pd.DataFrame({'x1':x1, 'x2':x2, 'y':y})
dataset.head()
```

```python
# split the dataset as train and test sets
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(dataset.drop('y', axis=1), dataset['y'], random_state = 42)
X_train.shape
```

```python
y_train.size
```

```python
# Train a decision tree regressor
from sklearn.tree import DecisionTreeRegressor
reg = DecisionTreeRegressor()
reg.fit(X_train, y_train)
y_pred = reg.predict(X_test)
y_pred.size
```

```python
# draw a plot check consistence between y_test and y_pred
# first zip y_test and y_pred
y_zip = list(zip(y_test, y_pred))
y_zip[0:3]
```

```python
# Sort the zip list of tuples based on y_test
y_sort = sorted(y_zip, key=lambda x: x[0], reverse =False)
y_sort[0:5]
```

```python
# Get sorted y_test and y_pred
y_test, y_pred = zip(*y_sort)
y_test[0:5]
```

```python
# plot y_test and y_pred
%matplotlib inline

import matplotlib.pyplot as plt
plt.plot(y_test, color='red')
plt.plot(y_pred, color='green')

# the predicted values are pretty consistent with the real y values
```

# Try with dataset with features with two if-then relationships

```python
# Let's create a dataset with five features
# when x4==1, x1*x2; when x4==-1,x1/x2
# when x5==1, +x3, when x5==-1, -x3
import numpy as np
import pandas as pd
np.random.seed(1)
size = 1000
x1 = np.random.random(size)
x2 = np.random.random(size)
x3 = np.random.random(size)

x4 = [np.random.choice([-1,1]) for _ in range(size)]
x5 = [np.random.choice([-1,1]) for _ in range(size)]

dataset = pd.DataFrame({'x1':x1, 'x2':x2, 'x3':x3, 'x4':x4, 'x5':x5})
```

```python
# dataset['y']=1
dataset.loc[(dataset.x4==1) & (dataset.x5==1),'y']= dataset.loc[(dataset.x4==1) & (dataset.x5==1),'x1']\
                                                *dataset.loc[(dataset.x4==1) & (dataset.x5==1),'x2']\
                                                +dataset.loc[(dataset.x4==1) & (dataset.x5==1),'x3']
dataset.loc[(dataset.x4==1) & (dataset.x5==-1),'y']= dataset.loc[(dataset.x4==1) & (dataset.x5==-1),'x1']\
                                                *dataset.loc[(dataset.x4==1) & (dataset.x5==-1),'x2']\
                                                -dataset.loc[(dataset.x4==1) & (dataset.x5==-1),'x3']
dataset.loc[(dataset.x4==-1) & (dataset.x5==1),'y']= dataset.loc[(dataset.x4==-1) & (dataset.x5==1),'x1']\
                                                /dataset.loc[(dataset.x4==-1) & (dataset.x5==1),'x2']\
                                                +dataset.loc[(dataset.x4==-1) & (dataset.x5==1),'x3']
dataset.loc[(dataset.x4==-1) & (dataset.x5==-1),'y']= dataset.loc[(dataset.x4==-1) & (dataset.x5==-1),'x1']\
                                                /dataset.loc[(dataset.x4==-1) & (dataset.x5==-1),'x2']\
                                                -dataset.loc[(dataset.x4==-1) & (dataset.x5==-1),'x3']

dataset['y']=dataset['y'] + np.random.random(size)/100 

dataset.head(20)
```

```python

# try with decision tree regressor first
X_train, X_test, y_train, y_test = train_test_split(dataset.drop('y', axis=1), dataset['y'], random_state = 42)

from sklearn.tree import DecisionTreeRegressor
reg = DecisionTreeRegressor()
reg.fit(X_train, y_train)
y_pred = reg.predict(X_test)
y_zip = list(zip(y_test, y_pred))
y_sort = sorted(y_zip, key=lambda x: x[0], reverse =False)
y_test, y_pred = zip(*y_sort)
plt.plot(y_test, color='red')
plt.plot(y_pred, color='green')

# the decision tree regressor work well
```

```python
# try with XGB regressor 
X_train, X_test, y_train, y_test = train_test_split(dataset.drop('y', axis=1), dataset['y'], random_state = 42)

from xgboost import XGBRegressor
reg = XGBRegressor()
reg.fit(X_train, y_train)
y_pred = reg.predict(X_test)
y_zip = list(zip(y_test, y_pred))
y_sort = sorted(y_zip, key=lambda x: x[0], reverse =False)
y_test, y_pred = zip(*y_sort)
plt.plot(y_test, color='red')
plt.plot(y_pred, color='green')

# the XGB regressor work better than decision tree regressor
```

```python
# try with SVM regressor with a linear kernel
X_train, X_test, y_train, y_test = train_test_split(dataset.drop('y', axis=1), dataset['y'], random_state = 42)

from sklearn.svm import SVR
reg = SVR(gamma='auto', kernel='linear')
reg.fit(X_train, y_train)
y_pred = reg.predict(X_test)
y_zip = list(zip(y_test, y_pred))
y_sort = sorted(y_zip, key=lambda x: x[0], reverse =False)
y_test, y_pred = zip(*y_sort)
plt.plot(y_test, color='red')
plt.plot(y_pred, color='green')
```
